```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('sales1.csv')
df2 = pd.read_csv('sales2.csv')
df3 = pd.read_csv('sales2.csv')
```

# These datsets were sourced from RTO Telangana

## Sales1, Sales2 and Sales3 include data of One Quarter

```python
df1.head()
```

```python
df2.head()
```

```python
print(df1.info(),df2.info(),df3.info())
```

```python
df = pd.concat((df1,df2,df3),axis=0)
```

```python
df['fuel'].value_counts()
```

```python
label = df['fuel'].value_counts().index
value = df['fuel'].value_counts().values.astype(float)
```

```python
import matplotlib.pyplot as plt
import seaborn
```

```python
plt.pie(value[:4],labels = label[:4],autopct='%0.2f%%')
plt.title('Vehicle fuel Type')
```

## Only 5 percent of the total vehicles purched are electric 

## Petrol vechiles take a major stake and dominate the market ( 80 percent)
